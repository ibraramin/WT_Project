```python
import pandas as pd


def clean_csv(input_filename="Recipeezz/dt.csv", output_filename="cleaned_recipes.csv"):
    """
    Reads a CSV file, removes its first two columns, and saves the result.

    Args:
        input_filename (str): The name of the CSV file to read.
        output_filename (str): The name for the cleaned CSV file.
    """
    try:
        # Read the CSV file into a pandas DataFrame
        df = pd.read_csv(input_filename)

        print(f"Original columns: {df.columns.tolist()}")
        if df.shape[0] > 0:
            print("First few rows of the original data:")
            print(df.head())
        else:
            print("The CSV file is empty or has no data rows.")
            # Save an empty dataframe with original headers if any, or just an empty file
            if df.shape[1] > 2:
                df.iloc[:, 2:].to_csv(output_filename, index=False)
            else:  # Not enough columns to drop two
                df.to_csv(output_filename, index=False)
            print(
                f"Saved an empty or minimally processed file as {output_filename} as there were not enough rows/columns for full processing."
            )
            return

        # Check if there are at least two columns to drop
        if df.shape[1] >= 2:
            # Drop the first two columns by their positional index
            # iloc[:, 2:] selects all rows and columns from the 3rd column onwards (index 2)
            df_cleaned = df.iloc[:, 2:]

            print(f"\nColumns after cleaning: {df_cleaned.columns.tolist()}")
            print("First few rows of the cleaned data:")
            print(df_cleaned.head())

            # Save the cleaned DataFrame to a new CSV file
            # index=False ensures that pandas doesn't write the DataFrame index as a new column
            df_cleaned.to_csv(output_filename, index=False)
            print(f"\nSuccessfully cleaned the CSV. Saved as '{output_filename}'")
        elif df.shape[1] < 2 and df.shape[1] > 0:
            # If there's only one column, we can't drop two.
            # We'll save it as is, or you might want to decide to save an empty file.
            print("\nThe CSV file has only one column. Cannot drop two columns.")
            print("Saving the file as is or an empty file if preferred.")
            # Option 1: Save as is
            # df.to_csv(output_filename, index=False)
            # Option 2: Save an empty dataframe if the intention is to always remove leading columns
            pd.DataFrame().to_csv(output_filename, index=False)  # Creates an empty CSV
            print(
                f"Saved an empty file as '{output_filename}' as there was only one column."
            )
        else:  # 0 columns
            print("\nThe CSV file has no columns to process.")
            pd.DataFrame().to_csv(output_filename, index=False)  # Creates an empty CSV
            print(f"Saved an empty file as '{output_filename}'.")

    except FileNotFoundError:
        print(
            f"Error: The file '{input_filename}' was not found. Make sure it's in the same directory as the script."
        )
    except pd.errors.EmptyDataError:
        print(f"Error: The file '{input_filename}' is empty and cannot be processed.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")


if __name__ == "__main__":
    # You can change the filenames here if needed
    input_csv_file = "Recipeezz/dt.csv"
    output_csv_file = "cleaned_recipes.csv"

    clean_csv(input_csv_file, output_csv_file)
```
